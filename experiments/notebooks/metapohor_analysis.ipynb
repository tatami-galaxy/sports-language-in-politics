```python
import json
```

```python
with open('/users/ujan/Downloads/exact_matches_politics_42_10000_v2.json') as f:
    pol_data = json.load(f)
```

```python
total = 0
no_list = ['out', 'up', 'tip', 'check']

sorted_meta_pol = dict(sorted(pol_data.items(), key=lambda item: item[1][0], reverse=True))
for key, val in sorted_meta_pol.items():
    if key in no_list:
        print(key, val[0])
        continue
    #print(key, val[0])
    total += val[0]
```
