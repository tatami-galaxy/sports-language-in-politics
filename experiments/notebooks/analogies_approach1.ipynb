```python
with open('/users/ujan/sports-language-in-politics/data/processed/token_dict_42_1000.json') as f:
    avg_data = json.load(f)

with open('/users/ujan/sports-language-in-politics/data/processed/comment_dict_42_1000.json') as f:
    comment_data = json.load(f)
```

```python
diff_dict = {'id':[], 'race':[], 'competition':[], 'championship':[], 'tournament':[]}
for id, val in comment_data.items():
    diff_dict['id'].append(id)
    ref = val['election']
    diff_dict['race'].append(ref - val['race'])
    diff_dict['competition'].append(ref - val['competition'])
    diff_dict['championship'].append(ref - val['championship'])
    diff_dict['tournament'].append(ref - val['tournament'])
```

```python
for i in range(len(diff_dict['id'])):
    if diff_dict['race'][i] < 0:
        print(data_df.filter(pl.col('id')==diff_dict['id'][i])['body'][0])
        print(' ')
```

```python
diff_dict = {'id':[], 'fan':[], 'fans':[], 'spectator':[], 'spectators':[]}
for id, val in comment_data.items():
    diff_dict['id'].append(id)
    ref = val['voters']
    diff_dict['fan'].append(ref - val['fan'])
    diff_dict['fans'].append(ref - val['fans'])
    diff_dict['spectator'].append(ref - val['spectator'])
    diff_dict['spectators'].append(ref - val['spectators'])
```

```python
for i in range(len(diff_dict['id'])):
    if diff_dict['fans'][i] < 0:
        print(data_df.filter(pl.col('id')==diff_dict['id'][i])['body'][0])
        print(' ')
```
